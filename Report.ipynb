### 2.1 functions used for data extraction and cleaning

The function generate_csv creates a csv file from a given url. The desired csv file name and parent folder are provided as arguments. Once the csv file is created it is stored in the specified parent folder.

Formula One teams usually change their names to accomodate new title sponsors at the end of a season. The update_teamNames function hence updates the old team names in the data to their current names. This is done so that the CSV files can be concatenated easily.

```python
#function that generates a csv from table in given link
def generate_csv(url, csv_name, parentfolder):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')

    #getting the table
    table = soup.find('table', {'class':'resultsarchive-table'})
    headers = []

    for i in table.find_all('th'):
        title = i.text.strip()
        headers.append(title)

    df = pd.DataFrame(columns = headers)

    for row in table.find_all('tr')[1:]:
        data = row.find_all('td')
        row_data = [td.text.strip() for td in data]
        length = len(df)
        df.loc[length] = row_data
        df.to_csv((parentfolder+csv_name), index=False, encoding='utf-8')
        
def update_teamNames(old_team_names, current_team_names, csv_name, parentfolder):
    df  = pd.read_csv(parentfolder + csv_name)
    newdf = df.replace(to_replace=old_team_names,value=current_team_names)
    #some teams changed names within a year and are hence listed twice in csv
    #we filter out these names to prevent double counting
    if len(newdf) > len(current_team_names):
        # Accessing the last row
        last_row = len(newdf)
        # Deleting the last row
        newdf = newdf.drop(newdf.index[last_row-1])
    newdf.to_csv(parentfolder + csv_name, index = False,encoding='utf-8')
```

## 4 Impact budget has on qualifying performance

This section will aim to determine whether there is any correlation between a teams budget and it's qualifying performance. A teams lap times will be compared to the average lap times of all the teams. Furthermore, a teams budget will be compared to the average budget of all the teams. 


### Qualifying Format

The outcome from qualifying determies the starting positon for the race. Qualifying takes place over three sessions called Q1, Q2 and Q3. In each session the cars are driven around the circuit to set the fastest possible lap time. The five slowest cars from each session will be eliminated. Therefore Q1 decides the starting positions in the race for positions 20-16. Q2 for 15-11 and Q3 for 10-1. Qualifying does not give the team any points, rather it decides the starting position in the race.

### Average of qualifying times

In order to evaluate the performace of a team against others, we compare a team's average lap times against the average lap times of the all the teams.



### 4.1 Functions used to extract data from csv

```python

#get seconds from lap time
def get_sec(time):
    time = time.replace(":",".")
    m, s, mil = time.split('.')
    return int(m) * 60 + int(s)

def get_grid_avg(csv):
    df = pd.read_csv(csv)
    #get list Q3 qulifying times
    laps = list(df['Q1'])+list(df['Q2'])+list(df['Q3'])
    # remove all empty and did not finish from list
    laps = [x for x in laps if str(x) != 'nan']
    laps = [x for x in laps if str(x) != 'DNF']
    tottimes = 0
    for i, item in enumerate(laps):
       laps[i] = get_sec(laps[i])
       tottimes += laps[i]
    return (tottimes/len(laps))


def get_team_avg(csv, team):
    df = pd.read_csv(csv)
        
    laptimes = df.loc[df.Car == team, ['Q1','Q2','Q3']].values.flatten().tolist()

    laptimes = [x for x in laptimes if str(x) != 'nan']
    laptimes = [x for x in laptimes if str(x) != 'DNF']
    tottimes = 0
    for i, lap in enumerate(laptimes):
       laptimes[i] = get_sec(laptimes[i])
       tottimes += laptimes[i]
    return (tottimes/(len(laptimes)))


def get_annual_avg_budget(year):
    
    budget = pd.read_csv('team_budgets.csv')
    
    #generate list by listing all team names
    teams = ['Mercedes','Ferrari','RedBull Racing','Mclaren','Alpine','Aston Martin','AlphaTauri','Alfa Romeo','Williams','Haas']

    budgets = budget.loc[budget.Year == year, teams].values.flatten().tolist()
    
    #remove all nan 
    budgets = [x for x in budgets if str(x) != 'nan']
    return sum(budgets) / len(budgets)

def get_team_budget(team):
    
    budget = pd.read_csv('team_budgets.csv')
    budgetList = list(budget[team])
    #remove all nan 
    budgetList = [x for x in budgetList if str(x) != 'nan']
    return budgetList
```

### 4.2 Functions used to plot graphs

```python
def plotTeamBudgetGraph(team):
    
    fig = plt.figure(figsize=(7,3))

    fig.suptitle( team + ' vs Average Budget', fontsize=15, weight="bold")
    
    years = [2015,2016,2017,2018,2019]
    
    
    team_budget = get_team_budget(team)
    avg_team_budget =[]
    
    for year in years:
        avg_team_budget.append(get_annual_avg_budget(year))
        
    plt.plot(years,avg_team_budget,'r.-',label = "Average Budget of All Teams")
    plt.plot(years,team_budget,'b.-', label = "Budget of "+team)
            
    leg = fig.legend(loc="upper left",fancybox=True, bbox_to_anchor=(0.3, 0.9))
   
    plt.xlabel('years')
    plt.xticks(years)
    plt.ylabel('Budget USD (millions)')
    plt.grid()
    
    
def poltLapTimesgraph(team,title,csv_list,ax):
    quali_fldr_path = 'qualifying times csvs\\'
    years = [2015,2016,2017,2018,2019]
    avg_laptimes =[]
    tm_avg_laptimes =[]
    
    for csv in csv_list:
        laptime = get_grid_avg(quali_fldr_path+csv)
        avg_laptimes.append(laptime)
        laptime = get_team_avg(quali_fldr_path+csv,team)
        tm_avg_laptimes.append(laptime)
    
    ax.plot(years,avg_laptimes,'r.-',label = "All Team Avg")
    ax.plot(years,tm_avg_laptimes,'b.-', label =team)
    
    leg = ax.legend(loc="upper right",fancybox=True)
   
    ax.set_title(title)
    ax.set_xlabel('Year')
    ax.set_ylabel('Time[s]')
    ax.grid()

#list holding the names of csv's to be called by the functions
sg_quali = ["sg_2015.csv","sg_2016.csv","sg_2017.csv","sg_2018.csv","sg_2019.csv",]
mon_quali = ["mon_2015.csv","mon_2016.csv","mon_2017.csv","mon_2018.csv","mon_2019.csv",]
suz_quali = ["suz_2015.csv","suz_2016.csv","suz_2017.csv","suz_2018.csv","suz_2019.csv",]
abu_quali = ["abu_2015.csv","abu_2016.csv","abu_2017.csv","abu_2018.csv","abu_2019.csv",]
    
```

## 6 Correlation between points and budget

In section will aim to determine whether there is any correlation between a teams budget and the points it scores in a season. Two charts will be plotted for each team. The first chart plots budget and points against time. The second chart plots a correlation between budget and time.


### 6.1 Defining functions for plotting graphs

The BudgetPointGraph function creates a graph that plots budget and points against time for a given team. It obtains budgets from the team_budgets.csv and points from the five_yr_pts dataframe defined previously.

The correlationPlot function creates a correlation plot for budget and points for a given team. Budgets are obtained from the team_budgets.csv and a team's points from he five_yr_pts dataframe.

```python

def BudgetPointGraph(team):
    
    budget = pd.read_csv('team_budgets.csv')
    team_budget = list(budget[team])
    
    points = five_yr_pts.loc[five_yr_pts.Team == team, ['2015','2016','2017','2018','2019']].values.flatten().tolist()
    
    fig, ax1 = plt.subplots()
    years_int = [2015,2016,2017,2018,2019]
    
    ax1.set_title(team)
    
    ax1.bar(years,team_budget,color='cornflowerblue',width=0.45)
    ax2 = ax1.twinx()
    ax2.plot(years,points,color='red',marker='.')

    ax1.set_ylabel('Budget Millions (USD)',color='blue',weight = 'bold')
    ax2.set_ylabel('Points',color='red', weight = 'bold')

    ax1.tick_params(axis='y',colors='blue')
    ax2.tick_params(axis='y',colors='red')

    plt.show()
    
    
def correlationPlot(team):
    
    budget = pd.read_csv('team_budgets.csv')
    team_budget = list(budget[team])
    
    points = five_yr_pts.loc[five_yr_pts.Team == team, ['2015','2016','2017','2018','2019']].values.flatten().tolist()

    # adds the title
    plt.title(team + ' correlation Points vs Budget')

    # plot the data
    plt.scatter(team_budget,points)
    
    # fits the best fitting line to the data
    plt.plot(np.unique(team_budget),
             np.poly1d(np.polyfit(team_budget, points, 1))
             (np.unique(team_budget)), color='red')

    # Labelling axes
    plt.xlabel('Budget USD millions')
    plt.ylabel('points')
    
```
